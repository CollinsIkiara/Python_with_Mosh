```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

music_data = pd.read_csv('music.csv')
X = music_data.drop(columns=['genre']) # Creates the input set (the input data the model will accept)
y = music_data['genre'] # Creates the output set (what the model is supposed to predict)

model = DecisionTreeClassifier() # Create an instance of our model
model.fit(X, y) # Trains our model on the input and output training data

# The below code creates a visualization of our model that we can view on VS Code.
tree.export_graphviz(model, out_file='music-recommender.dot',
                    feature_names=['age', 'gender'], # So that we can see the rules in our nodes
                    class_names=sorted(y.unique()), # Displays the unique class names (y-variable) in alphabetical order
                    label='all', # Ensures all our nodes are labelled
                    rounded=True, # Makes our nodes have rounded corners
                    filled=True) # Fills our nodes with colour
```
